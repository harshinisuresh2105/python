```python


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data={
   'month':["Jan","Feb","Mar","Apr","May","Jun","Jul"],
   'expenses':[7000 ,4000 ,9000 ,6000 ,2000 ,4500 ,3000],
   'categories': ["Food", "Travel", "Shopping", "Bills", "Entertainment", "Savings","Medical"],
   }
df=pd.DataFrame(data)
print(df)
```

```python
#Line chart

import matplotlib.pyplot as plt
df.plot(x="month",y="expenses", marker="o",color="orange")
plt.title("Monthly Expenses Trend")
plt.xlabel("month")
plt.ylabel("expenses")
plt.show()
```

```python
#seaborn linechart

sns.lineplot(data=df,x="month",y="expenses",marker="o",color="red")
plt.title("Monthly Expenses")
plt.show()
```

```python
#Bar plot
import matplotlib.pyplot as plt
df.plot(x="categories", y= "expenses", kind="bar", color="green")
plt.title("Expenses trend by categories")
plt.xlabel("categories")
plt.ylabel("expenses")
plt.show()
```

```python
sns.barplot(data=df,x="categories",y="expenses",color="pink")
plt.title("Categorical Expenses")
plt.show()
```

```python
#pie chart using matplotlib

plt.pie(df["expenses"], labels=df["categories"], autopct='%1.1f%%')
plt.title("Monthly Expense Breakdown")
plt.show()
```
